### iii. Outliers

71,70,90,70,70,60,70,72,72,320,71,69

We can do the following

1. First we will sort the data

60,69,70,70,70,70,71,71,72,72,90,320

2. Next we'll calculate the median. The median is the middle data after sorting

The median is (70 + 71)/2 = 71.5

3. Then we'll calculate the lower quartile(Q1).Q1 is the middle value of the first half of the dataset

First half of the dataset : 60,69,70,70,70,70

The average is (70 + 70)/2 = 70

Q1 = 70

4. Then we calculate the upper quartile (Q3)

Q3 is the middle value of the second half of the dataset

Second half of the dataset : 71,71,72,72,90,320

The average is (72 + 72)/2 = 72

5. Then we find the inter quartile range (IQR)

IQR = Q3 - Q1

IQR = 72 - 70 = 2

6. Next we find the upper and lower fences.

Lower fence = Q1 - 1.5(IQR) = 70 - 1.5(2) = 67
Upper fence = Q3 + 1.5(IQR) = 72 + 1.5(2) = 75

Any points lower than lower fence and greater than upper fence are outliers, in our case the outliers are 90,60,320

### ii. Label Encoding

This is a technique inwhich we replace each value in a categorical column with numbers from 0 to N-1. For example, say we've got a list of employee names in a column. After performing label encoding each employee name will be assigned a numeric label. But this might not be suitable for all cases as the model might consider the numeric values to be the weights assigned to the data. 

Therefore label encoding works best with ordinal data.

### iii. One Hot Encoding

In label encoding categorical data is converted to numerical data, and the values are assigned labels (1,2 and 3). Predictive models which use this numerical data for analysis might sometimes mistake these labels for some kind of order (for example, a model might think that a label 3 is better than a label 1, which is incorrect. In order to avoid this confusion we use one-hot encoding. Here, the label-encoded data is further divided into n number of columns. Here n denotes the total number of unique labels generated while performing label encoding. For example, say that 3 new labels are generated through label encoding, the columns will be divided into 3 parts. So, the value of n is 3.

# Topic 6: Data Discretization

Data discretization is the process of converting continuous data into discrete buckets by grouping it. Training a model with discrete data becomes more faster and effective than when attempting the same with continuous data.
Although continuous-valued data contains more information, huge amounts of data can slow the model down. Discretization can help us strike a balance between both.
Some famous methods of data discretization are binning and using a histogram. Although data discretization is useful, we need to effectively pick the range of bucket which is a challenge.

The main challenge is to choose the number of intervals or bins and how to decide on their width.

We can use the function pandas.cut(), this function is used for bucketing and sorting of augmented data.
